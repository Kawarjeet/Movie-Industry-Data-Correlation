```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

```

```python
#reading data 
df = pd.read_csv(r'C:\Users\18572\Downloads\movies.csv')
```

```python
#looking at the dataset
df.head()
```

```python
#checking for missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#data types query
print(df.dtypes)
```

```python
#performing some basic data cleaning in the following queries
```

```python
#changing the data types of the 'budget' and 'gross' columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False) #sorting the data by 'gross' collection in descending order
```

```python
pd.set_option('display.max_rows', None) #query to view all of the data in the dataset
```

```python
#dropping duplicate company names
df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
#dropping any duplicate data
df.drop_duplicates().head()
```

```python
#scatter plot with budget and gross 
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#plotting budget vs gross using seaborn for a better correlation between them
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"}) #regression plot
```

```python
#finding correlation 
df.corr() #by default pearon correlation is being used. We may also use other correlations like kendall and spearman
```

```python
df.corr(method='spearman')
```

```python
df.corr(method='kendall')
```

```python
# using heatmap for the correlation matrix. Brighter color are highly correlated 
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# assigning numeric values to columns with string values
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#unstacking the correlation matrix
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs.head()
```
